The goal of this project is to take a look at audience reviews for Marvel's new movie Shang Chi and develop a topic model that would help Disney understand the key topics that audiences are talking about when it comes to the new film. The hope would to use this model to provide insight on what aspects of future movies can be improved or reinforce that the quality of the movie is appealing to audiences. 

In order to achieve this goal, ~2,000 reviews were pulled from google reviews and went through a series of pre processing steps in order to make them ready for NLP. Since the reviews are all user inputs with no strict rules for grammar and spelling, all reviews had punctuation removed and spelling checked using textblob. Words were also lemmatized to avoid similar words being counted multiple times. Lastly, the movie title and actors names were compounded.  

In order to shed light on some of the possible topics in reviews, I created two topic models using both LSA as well as NMF. My initial findings are not too promising as a large majority of the reviews are positive and generally talk about very similar topics. In order to obtain lists of words that seem somewhat distinct I set the max_df for both models to 0.2 since anything higher had words like marvel,movie, and good in almost all of my topics.
